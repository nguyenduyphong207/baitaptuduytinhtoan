```python
#W4A6
def prime(x):
    if x < 2:
        return False
    for i in range(2, int(x**0.5) + 1):
        if x % i == 0:
            return False
    return True

a = int(input("Nhập số nguyên dương a: "))
b = int(input("Nhập số nguyên dương b: "))
sum = 0
for num in range(a, b + 1):
    if prime(num):
        sum += num
print(sum)
```

```python
#w4a9
def check(x):
    for j in range(len(x)):
        for k in range(j + 1, len(x)):
            if x[j] == x[k]:
                return False
    return True
n = int(input())
List_print = []

for i in range(1, n+1):
    if i ** 0.5 == int(i ** 0.5):
        if check(list(str(i))):
            List_print.append(i)
if len(List_print) == 0:
    print("no number")
else:
    print(*List_print)
        
```

```python
#w4a10
def collatz(n):
    lenght = [n]
    while n != 1:
        n = n//2 if n%2==0 else 3*n+1
        lenght.append(n)
    return lenght

n = int(input())
find_max = []
find_x_min = []

for y in range(1, n+1):
    find_max.append(len(collatz(y)))
a = max(find_max)

for x in range(1, n+1):
    if len(collatz(x)) == a:
        find_x_min.append(x)

print(min(find_x_min))
print(a)
```

```python
#w4a26
def fibo(x):
    fibo_list = []
    for i in range(x):
        if i == 0 or i == 1:
            fibo_list.append(1)
        else:
            fibo_list.append(int(fibo_list[i-1]) + int(fibo_list[i-2]))
    return fibo_list   

A = int(input())
x = 1

while max(fibo(x)) <= A:
    x += 1
print(max(fibo(x-1)))
```

```python
#w4a32
password = input()
dai = len(password)
dac_biet = 0
in_hoa = 0
in_thuong = 0
chu_so = 0
for i in password:
    if not i.isalnum():
        dac_biet += 1
    elif i.isupper():
        in_hoa += 1
    elif i.islower():
        in_thuong += 1
    elif i.isdigit():
        chu_so += 1

if dai > 6 and dac_biet >= 1 and in_hoa >= 1 and in_thuong >= 1 and chu_so >= 1:
    print("mạnh")
else:
    print("yếu")
```
