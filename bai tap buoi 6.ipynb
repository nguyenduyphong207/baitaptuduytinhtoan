```python
#W6A9
keys1=input().split()
values1=map(int,input().split())
d1=dict(zip(keys1,values1))

keys2=input().split()
values2=map(int,input().split())
d2=dict(zip(keys2,values2))
result={}
for key in d1:
  if key in d2:
    result[key]=d1[key]+d2[key]
  else:
      result[key]=d1[key]
for key in d2:
    if key not in d1:
      result[key]=d2[key]

print(dict(sorted(result.items())))



```
