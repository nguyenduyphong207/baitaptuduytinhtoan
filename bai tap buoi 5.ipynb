```python
#w5a3
def prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

prime(11)
```

```python
#w5a4
def perfect(n):
    sum = 0
    for i in range(1, n):
        if n % i == 0:
            sum += i
    if sum == n:
        return True
    else:
        return False
    
perfect(28)
```

```python
#w5a5
list_num = [1, 2, 3, 4, 5, -6, 7, 8, 19, 1, 2, 3, 4, 5]

def find_positive():
    global list_num
    for i in range(len(list_num)):
        if list_num[i] > 0:
            print(i)
            break
        else:
            print(-1)

find_positive()
            
```

```python
#w5a6
def factorial(n):
    result = 1
    if n == 0 or n == 1:
        return 1
    else:
        for i in range(2, n + 1):
            result *= i
        return result
    
factorial(5)
```

```python
#w5a7
def calc(a, op, b):
    if op == '+':
        return a + b
    elif op == '-':
        return a - b
    elif op == '*':
        return a * b
    elif op == '/':
        return a / b
    else:
        return False
    
a, op, b = map(str, input().split())
a = int(a)
b = int(b)
print(calc(a, op, b))
```

```python
#w5a8
def hamming(x, y):
    a = x.bit_length()
    b = y.bit_length()
    if a > b:
        distance = a - b
    elif a < b:
        distance = b - a
    else:
        distance = 0
    print(distance)

hamming(1, 4)
```

```python
#w5a9
def sum_digits(n):
    sum = 0
    while n > 0:
        a = n % 10
        sum += a
        n = n // 10
    return sum

sum_digits(1234)
```
