```python
#W7A1
def binary_search(arr,target):
  left,right=0,len(arr)-1
  while left <= right:
    mid=(right+left)//2
    if arr[mid]==target:
      return mid
    elif arr[mid]<target:
      left=mid+1
    else:
      right=mid-1
  return arr
print(binary_search([1,2,3,5,8],5))
```

```python
#W7A3
def bubble_sort(arr):
  n=len(arr)
  for i in range (n):
    swapped=False
    for j in range(0,n-i-1):
      if arr[j]>arr[j+1]:
        arr[j],arr[j+1]=arr[j+1],arr[j]
        swapped=True
    if not swapped:
      break
  return arr
def selection_sort(arr):
  n=len(arr)
  for i in range(n):
    min_idx=i
    for j in range(i+1,n):
      if arr[j]<arr[min_idx]:
        min_idx=j
    arr[i],arr[min_idx]=arr[min_idx],arr[i]
  return arr
print(bubble_sort([64,34,25,12,22,11,90]))
print(selection_sort([64,34,25,12,22,11,90]))

```

```python
#W7A8
import heapq

def minInterval(intervals, queries):
    # Sắp xếp intervals theo l
    intervals.sort(key=lambda x: x[0])

    # Gắn chỉ số gốc cho queries rồi sắp xếp
    qs = sorted((q, i) for i, q in enumerate(queries))

    res = [-1] * len(queries)
    heap = []  # (length, r)
    i = 0      # con trỏ intervals

    for q, idx in qs:
        # Thêm các interval có l <= q
        while i < len(intervals) and intervals[i][0] <= q:
            l, r = intervals[i]
            heapq.heappush(heap, (r - l + 1, r))
            i += 1

        # Loại các interval không còn chứa q
        while heap and heap[0][1] < q:
            heapq.heappop(heap)

        # Kết quả cho query q
        if heap:
            res[idx] = heap[0][0]
        else:
            res[idx] = -1

    return res
print(minInterval([[1,4],[2,4],[3,6],[4,4]],[2,3,4,5]))
```
